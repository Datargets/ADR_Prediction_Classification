```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# Define three different confusion matrices
confusion_matrices = [
    np.array([
        [324, 2, 3],
        [5, 443, 4],
        [4, 9, 213]
    ]),
    np.array([
        [326, 0, 3],
        [4, 447, 1],
        [3, 6, 217]
    ]),
    np.array([
        [324, 3, 2],
        [4, 446, 2],
        [3, 4, 219]
    ]),
    np.array([
        [267, 28, 33],
        [27, 387, 38],
        [46, 33, 147]
    ])
]

# Define the labels for the confusion matrices
labels = ['Diabetes', 'Cancer', 'High_Bp']

# Create a custom colormap from white to light gray
cmap = mcolors.LinearSegmentedColormap.from_list("light_gray", ["white", "lightgray"])

# Create the figure with 3 subplots
fig, axes = plt.subplots(1, 4, figsize=(14, 3))

# Titles for the subplots
titles = ['Bert Fine-Tune with LoRA', 'DistilBert Fine-Tune with LoRA', 'RoBerta Fine-Tune with LoRA', 'Bloom Fine-Tune with LoRA']

# Loop through each subplot and corresponding confusion matrix
for i, (conf_matrix, ax) in enumerate(zip(confusion_matrices, axes)):
    sns.heatmap(conf_matrix, annot=True, fmt="d", cmap=cmap,
                xticklabels=labels, yticklabels=labels, ax=ax)
    ax.set_title(titles[i])
    ax.set_xlabel('Predicted label')
    ax.set_ylabel('True label')

plt.tight_layout()
plt.show()
```
